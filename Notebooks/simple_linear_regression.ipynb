## Linear regression python 

Linear Regression with two variables in one dimensional data

```python
import numpy as np
import random
```

```python
n=10
x=np.arange(0,10,0.1)
y=3*x+2
n=x.shape[0]
max_itr=1000
alpha=0.0001
```

```python
# F(x)= t[0]*x + t[1]
def F(x, t):
    return x*t[0]+t[1]

def cost(y_est, y):
    E=y_est-y
    return E, np.linalg.norm(E,2)

def gradient(E,x):
    return np.array([2* np.matmul(E.T,x), 2*np.sum(E)])
```

```python
t=np.random.random(2)
```

```python
def fit(t, x, y, alpha, max_itr):
    for i in range(max_itr):
        print(t)
        y_est=F(x,t)
        E, c= cost(y_est, y)
        tg=gradient(E,x)
        t=t - alpha * tg
        
    return t
```

```python
fit(t, x, y, alpha, max_itr)
```
