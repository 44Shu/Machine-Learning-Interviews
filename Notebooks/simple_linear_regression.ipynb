## Linear regression python one dimensional data

Linear Regression with two variables in one dimensional data

 
 
 $$ F(x)= w_0 \times x + w_1 $$
 $$ C=|| F(x) - y ||_2^2 $$

```python
import numpy as np
import random
```

```python
n=10
x=np.arange(0,10,0.1)
y=3*x+2
n=x.shape[0]
max_itr=1000
alpha=0.0001
w=np.random.random(2)
```

```python
# F(x)= w[0]*x + w[1]
def F(x, w):
    return x*w[0]+w[1]

def cost(y_est, y):
    E=y_est-y
    return E, np.linalg.norm(E,2)

def gradient(E,x):
    return np.array([2* np.matmul(E.T,x), 2*np.sum(E)])
```

```python
def fit(w, x, y, alpha, max_itr):
    for i in range(max_itr):
        print(w)
        y_est=F(x,w)
        E, c= cost(y_est, y)
        wg=gradient(E,x)
        w=w - alpha * wg
        
    return w
```

```python
w = fit(w, x, y, alpha, max_itr)
```
